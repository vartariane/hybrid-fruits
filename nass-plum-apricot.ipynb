```python
ag_cleaned = ag.drop(columns=['Program', 'Period', 'State ANSI', 'Ag District Code', 'County ANSI', 'Commodity', 'CV (%)'])
ag_cleaned.head()
```

```python
ag_cleaned['Data Item'] = ag_cleaned['Data Item'].str.replace('PLUM-APRICOT HYBRIDS, INCL PLUMCOTS & PLUOTS - ', '')
```

```python
ag_us_total = ag_cleaned.query('`Geo Level` == "NATIONAL" and Domain == "TOTAL"')
ag_us_total.head(30)
```

```python
ag_us_total['Value'] = ag_us_total['Value'].str.replace(',', '')
```

```python
ag_us_total['Value'] = ag_us_total['Value'].astype('Int64')
```

```python
ag_us_total_all = ag_us_total['Data Item'].isin(['ACRES BEARING & NON-BEARING', 'OPERATIONS WITH AREA BEARING & NON-BEARING'])
```

```python
ag_us_total_all['Value'].plot(kind='bar')
```
